```python
import numpy as np
import pandas as pd
import material as ml
import thinfilm as tm
import matplotlib.pyplot as plt

void = ml.Non_Dispersion(1)
SG = pd.read_csv('spectrum/Nanya/Nanya_SG_T_0.csv')['SCAN']
layer0 = pd.read_csv('spectrum/Nanya/2IM/layer0.csv')['SCAN']
sub = ml.open_material('sodalime_Nanya','Material')

wl = np.linspace(400,700,301)

test = tm.Design(
    [void, sub],
    [None, None]
)

plt.plot(wl, SG, '-b',label = 'strengthened glass')
plt.plot(wl, layer0, '-k',label = 'glass')
plt.plot(wl, 100*tm.sec_transmit(test, wl), '--r',label = 'simulation')
plt.legend(loc = 'best')
plt.xlabel('wavelength (nm)')
plt.ylabel('transmittance')
plt.savefig('Nanya_substrate.png')
plt.show()
```
