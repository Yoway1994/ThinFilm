```python
import numpy as np
import pandas as pd
import material as ml
import thinfilm as tm
import matplotlib.pyplot as plt

air = ml.Non_Dispersion(1)
SiO2 = ml.open_material('SiO2_Nanya', 'Material')
NB = ml.open_material('Nb2O5_Nanya', 'Material')
ITO2 = ml.open_material('ITO_Nanya', 'Material')
ITO1 = ml.open_material('ITO1_Nanya', 'Material')
OC = ml.open_material('OC_Nanya', 'Material')
sodalime_Nanya = ml.open_material('sodalime_Nanya', 'Material')

file = pd.read_csv('./spectrum/Nanya/Bonsai_Macleod_R.csv')
tra_0 = file['tra_0'].values
dia_0 = file['dia_0'].values
bri_0 = file['bri_0'].values
tra_5 = file['tra_5'].values
dia_5 = file['dia_5'].values
bri_5 = file['bri_5'].values
tra_30 = file['tra_30'].values
dia_30 = file['dia_30'].values
bri_30 = file['bri_30'].values
wl = np.linspace(400, 700, 61)

d = [76, 30, 1465, 75, 45.3, 7.8]

bri = lambda th:tm.Design(
    [air, SiO2, ITO2, OC, ITO1, SiO2, NB, sodalime_Nanya],
    [None, th[0], th[1], th[2], th[3], th[4], th[5], None]
)

bri_Nanya = lambda th:tm.Design(
    [air, SiO2, ITO2, ITO1, SiO2, NB, sodalime_Nanya],
    [None, th[0], th[1], th[3], th[4], th[5], None]
)

dia = lambda th:tm.Design(
    [air, SiO2, ITO2, SiO2, NB, sodalime_Nanya],
    [None, th[0], th[1], th[4], th[5], None]
)

tra = lambda th:tm.Design(
    [air, SiO2, SiO2, NB, sodalime_Nanya],
    [None, th[0], th[4], th[5], None]
)

plt.plot(wl, 100*tm.sec_reflec(dia(d), wl, 0), '-k')
plt.plot(wl, dia_0,'ko', markersize = 3.5)
plt.plot(wl, 100*tm.sec_reflec(tra(d), wl, 0), '-b')
plt.plot(wl, tra_0,'bo', markersize = 3.5)
#plt.plot(wl, 100*tm.sec_reflec(bri(d), wl, 0), '-r')
#plt.plot(wl, bri_0,'ro', markersize = 3.5)
plt.show()
plt.plot(wl, 100*tm.sec_reflec(dia(d), wl, 5), '-k')
plt.plot(wl, dia_5,'ko', markersize = 3.5)
plt.plot(wl, 100*tm.sec_reflec(tra(d), wl, 5), '-b')
plt.plot(wl, tra_5,'bo', markersize = 3.5)
#plt.plot(wl, 100*tm.sec_reflec(bri(d), wl, 0), '-r')
#plt.plot(wl, bri_0,'ro', markersize = 3.5)
plt.show()
plt.plot(wl, 100*tm.sec_reflec(dia(d), wl, 30), '-k')
plt.plot(wl, dia_30,'ko', markersize = 3.5)
plt.plot(wl, 100*tm.sec_reflec(tra(d), wl, 30), '-b')
plt.plot(wl, tra_30,'bo', markersize = 3.5)
#plt.plot(wl, 100*tm.sec_reflec(bri(d), wl, 0), '-r')
#plt.plot(wl, bri_0,'ro', markersize = 3.5)
plt.show()
```
