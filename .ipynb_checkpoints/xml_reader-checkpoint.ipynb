```python
import xml.etree.ElementTree as ET
import numpy as np

tree = ET.parse('.\SE_material\OC_k.xml')
root = tree.getroot()

wl = []
n = []

for i in root.iter('point'):
    wl.append(float(i.attrib['dx']))
    n.append(float(i.attrib['dy']))
wn = np.array(wl)
nn = np.array(n)
print(wn, nn)
```
