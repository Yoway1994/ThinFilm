```python
import numpy as np
import pandas as pd
import material as ml
import thinfilm as tm
import matplotlib.pyplot as plt
```

```python
iton = pd.read_csv('./TSF/TSF_ITOn.csv')
itok = pd.read_csv('./TSF/TSF_ITOk.csv')
bk7 = pd.read_csv('./TSF/BK7.csv')
bk7k = pd.read_csv('./TSF/BK7k.csv')

ITOn_w = iton['nm'].values
ITOn_n = iton['n'].values
ITOk_w = itok['nm'].values
ITOk_k = itok['k'].values
BK7_w = bk7['wl'].values
BK7_n = bk7['n'].values
BK7k_w = bk7k['wl'].values
BK7k_k = bk7k['k'].values


ITO = ml.Material(ITOn_w, ITOn_n, ITOk_k, ITOk_w)
BK7 = ml.Material(BK7_w, BK7_n, BK7k_k, BK7k_w)
```

```python
air = ml.Non_Dispersion(1)

wl = np.linspace(400, 1800, 1401)
plt.plot(wl, ITO.nvalues(wl))
plt.show()
plt.plot(wl, ITO.kvalues(wl))
plt.show()
plt.plot(wl, BK7.nvalues(wl))
plt.show()
plt.plot(wl, BK7.kvalues(wl))
plt.show()
```

```python
t_R = pd.read_csv('./TSF/TSF_R.csv')['R']
t_w = pd.read_csv('./TSF/TSF_R.csv')['nm']
RaU = tm.Design(
    [air, ITO, BK7],
    [None, 300, None]
).reflectance(wl)
RaD = tm.Design(
    [BK7, ITO, air],
    [None, 300, None]
).reflectance(wl)

RbU = tm.Design(
    [BK7, air],
    [None, None]
).reflectance(wl)

TaU = tm.Design(
    [air, ITO, BK7],
    [None, 300, None]
).transmittance(wl)

R = (RaU+RbU*(TaU**2-RaU*RaD))/(1-RaD*RbU)

plt.plot(t_w, t_R, '-k',label = 'paper reference')
plt.plot(wl, R, '--r', label = 'with 2nd reflection')
plt.plot(wl, RaU, '--b',label = 'origin model')

plt.legend(loc = 'best')
#plt.savefig('2nd.png')
plt.show()
```
