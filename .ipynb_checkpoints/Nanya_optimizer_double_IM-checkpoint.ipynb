```python
import numpy as np
import pandas as pd
import material as ml
import thinfilm as tm
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution, minimize
```

```python
wl = np.linspace(400, 700, 301)
w = pd.read_csv('./spectrum/TEO/w.csv')['nm'].values

NB_1 = pd.read_csv('./spectrum/TEO/NB_1.csv')
NBn = NB_1['n'].values
NBk = NB_1['k'].values
NB = ml.Material(w, NBn, NBk, w)

SiO2_1 = pd.read_csv('./spectrum/TEO/SiO2_1.csv')
SiO21n = SiO2_1['n'].values
SiO21k = SiO2_1['k'].values
SiO2 = ml.Material(w, SiO21n, SiO21k, w)

ITO2 = pd.read_csv('./spectrum/TEO/ITO.csv')
ITOn = ITO2['n2'].values
ITOk = ITO2['k2'].values
ITO = ml.Material(w, ITOn, ITOk, w)

sub =  pd.read_csv('./spectrum/TEO/sodalime.csv')
soda_n = sub['n'].values
soda_k = sub['k'].values
sodalime = ml.Material(w, soda_n, soda_k, w)
#sodalime = ml.Non_Dispersion(1.6)
ITO1 = ml.open_material('ITO1_Nanya', 'Material')
OC = ml.open_material('OC_Nanya', 'Material')
air = ml.Non_Dispersion(1)

bri = lambda th:tm.Design(
    [air, OC, NB, SiO2, ITO, OC, ITO1, SiO2, NB, sodalime],
    [None, th[5]-1500, th[0], th[1], th[2], 1500, 58, th[3], th[4], None]
).reflectance(wl)

dia = lambda th:tm.Design(
    [air, OC, NB, SiO2, ITO, SiO2, NB, sodalime],
    [None, th[5], th[0], th[1], th[2], th[3], th[4], None]
).reflectance(wl)

tra = lambda th:tm.Design(
    [air, OC, NB, SiO2, SiO2, NB, sodalime],
    [None, th[5], th[0], th[1], th[3], th[4], None]
).reflectance(wl)

delta_d_t = lambda th: np.mean(abs(dia(th)-tra(th)))
delta_d_b = lambda th: np.mean(abs(dia(th)-bri(th)))
fom = lambda th: delta_d_t(th)
```

```python
sol = differential_evolution(
    delta_d_b, [
        (0,100),(75, 75),(30,30),(45,45),(8,8),(1500, 2500)
    ], disp=True, polish=False, popsize=20, mutation=0.3, recombination =0.9,strategy = 'best1exp'
)
print(sol.x)
```

```python
m1 = [0,75,30,45,8]
m2 = [6,28.7,30,24.7,8,2000]
m3 = [14.5,67.6,30.,32, 9, 2000]
m4 = [21,75.,30.,45.,9,1500]
tk = m4
plt.plot(wl, dia(tk), label = 'xyITO')
plt.plot(wl, tra(tk), label = 'spacing pattern')
plt.plot(wl, bri(tk), label = 'bridge')
plt.legend(loc = 'best')
plt.show()
#plt.savefig('ff.png')
```
