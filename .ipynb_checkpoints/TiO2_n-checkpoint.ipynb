```python
import numpy as np
import matplotlib.pyplot as plt
wavelength = [411, 426.72, 444.34, 464.21, 488.98, 516.26, 548.05, 586.25, 628.18, 680, 742.65, 912.35, 1042.06, 1203, 1442.75]
TiO2_n = [2.34786, 2.33138, 2.31109, 2.28202, 2.24594, 2.21702, 2.19754, 2.18272, 2.17144, 2.16102, 2.15008, 2.1045, 2.10095, 2.09569, 2.0885]
```

```python
# preprocessing
w_micron = np.array(wavelength)/1000
# training data
target_x = w_micron**2
target_y = np.array(TiO2_n)**2
# sample number
m = np.size(target_x)
# initial value
A = 4.2
B = 0.17
C = 0
plt.plot(target_x, (A + B/(target_x-C)))
plt.plot(target_x, target_y, '-s')
plt.show()
# cost function
testi = []
testJ = []
# learning rate
alpha = 0.1
# iteration
for i in range(1000):
    # hypothesis
    hofx = A + B/(target_x-C)
    # gradient descent
    Grad_J = 1/m*sum((hofx-target_y)*target_x.T)
    A = A - alpha* Grad_J
    B = B - alpha* Grad_J
    C = C - alpha* Grad_J
    # cost function
    J = 1/(2*m)*sum((hofx-target_y)**2)
    testi.append(i)
    testJ.append(J)
    if J < 0.0009:
        break
# plot figure
plt.plot(testi, testJ, 'r')
plt.show()
print('cost =', testJ[-1])
plt.plot(target_x, (A + B/(target_x-C)))
plt.plot(target_x, target_y, '-s')
plt.show()
print(A, B, C)
```

```python
# preprocessing
def nor(w):
    return [10**5.8/i**2 for i in w]
# initial value
A = 15
B = 16
C = 1
# sample number
m = np.size(wavelength)
# training data
target_x = np.array(nor(wavelength))
target_y = np.array(TiO2_n)**2
# cost function
testi = []
testJ = []
# learning rate
alpha = 0.01
for i in range(100):
    hofx = A + B*(target_x-C)
    Grad_J = 1/m*sum((hofx-target_y)*target_x.T)
    J = 1/(2*m)*sum((hofx-target_y)**2)
    A = A - alpha* Grad_J
    B = B - alpha* Grad_J
    C = C - alpha* Grad_J
    testi.append(i)
    testJ.append(J)
plt.plot(testi, testJ)
plt.show()
print('cost =', testJ[-1])
plt.plot(wavelength, np.sqrt(A + B*(target_x-C)))
plt.plot(wavelength, TiO2_n, 's')
plt.show()
print(A, B, C)
```
