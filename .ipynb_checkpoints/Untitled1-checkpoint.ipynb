```python
import numpy as np
import pandas as pd
import material as ml
import thinfilm as tm
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution, minimize
```

```python
sub = ml.open_material('Nanya_sodalime', 'Material')
wl = np.linspace(400,700,301)

void = ml.Non_Dispersion(1)
ws = pd.read_csv('./spectrum/TEO/ws.csv')['nm'].values
NB_file = pd.read_csv('./spectrum/TEO/NB.csv')
NB1n = NB_file['n1'].values
NB1k = NB_file['k1'].values
NB1 = ml.Material(ws, NB1n, NB1k, ws)

NB2n = NB_file['n2'].values
NB2k = NB_file['k2'].values
NB2 = ml.Material(ws, NB2n, NB2k, ws)

SiO2_1 = pd.read_csv('./spectrum/TEO/SiO2.csv')
SiO21n = SiO2_1['n'].values
SiO21k = SiO2_1['k'].values
SiO2 = ml.Material(ws, SiO21n, SiO21k, ws)

SiO2_2 = pd.read_csv('./spectrum/TEO/SiO2_wafer.csv')
SiO22n = SiO2_2['n']
SiO22k = SiO2_2['k']
SiO22w = SiO2_2['nm']
SiO2_wf = ml.Material(SiO22w, SiO22n, SiO22k, SiO22w)
```

```python
L1_hypo1 = tm.Design(
    [void, NB1, sub],
    [None, 3.6, None]
).reflectance(wl, 5)

L1_hypo2 = tm.Design(
    [void, NB2, sub],
    [None, 3.6, None]
).reflectance(wl, 5)

L3_hypo1 = tm.Design(
    [void, NB1, sub],
    [None, 13.8, None]
).reflectance(wl, 5)

L3_hypo2 = tm.Design(
    [void, NB2, sub],
    [None, 13.8, None]
).reflectance(wl, 5)
```

```python
ToS2 = pd.read_csv('./spectrum/ToS2/ToS2.csv')['SCAN'].values
L1 = pd.read_csv('./spectrum/ToS2/ToS2_L1.csv')['SCAN'].values
L3 = pd.read_csv('./spectrum/ToS2/ToS2_L3.csv')['SCAN'].values
plt.plot(wl, L1, 'ks', markersize = 1)
plt.plot(wl, 100*L1_hypo1)
plt.plot(wl, 100*L1_hypo2)
print(100*np.mean(abs(L1_hypo1-L1_hypo2)))
plt.show()
plt.plot(wl, L3, 'ks', markersize = 1)
plt.plot(wl, 100*L3_hypo1)
plt.plot(wl, 100*L3_hypo2)
print(100*np.mean(abs(L3_hypo1-L3_hypo2)))
plt.show()
```

```python
L2_hypo = tm.Design(
    [void, SiO2, sub],
    [None, 55.6, None]
).reflectance(wl, 5)

L2_hypo2 = tm.Design(
    [void, SiO2_wf, sub],
    [None, 55.6, None]
).reflectance(wl, 5)

L4_hypo = tm.Design(
    [void, SiO2, sub],
    [None, 23.8, None]
).reflectance(wl, 5)

L4_hypo2 = tm.Design(
    [void, SiO2_wf, sub],
    [None, 23.8, None]
).reflectance(wl, 5)

sub_hypo = tm.Design(
    [void, sub],
    [None, None]
).reflectance(wl, 5)
```

```python
plt.plot(wl, SiO2.nvalues(wl))
plt.plot(wl, SiO2_wf.nvalues(wl))
plt.show()
```

```python
L2 = pd.read_csv('./spectrum/ToS2/ToS2_L2.csv')['SCAN'].values
L4 = pd.read_csv('./spectrum/ToS2/ToS2_L4.csv')['SCAN'].values

plt.plot(wl, L2, 'ks', markersize = 1)
plt.plot(wl, 100*L2_hypo)
plt.plot(wl, 100*L2_hypo2)
plt.plot(wl, 100*sub_hypo, '--k')
plt.show()
plt.plot(wl, L4, 'ks', markersize = 1)
plt.plot(wl, 100*L4_hypo)
plt.plot(wl, 100*L4_hypo2)
plt.plot(wl, 100*sub_hypo, '--k')
plt.show()
```

```python
file = pd.read_csv('./spectrum/Nanya_ITO_grading.csv')
w = file['nm'].values

ITOs = pd.read_csv('./spectrum/TEO/LT_ITO_1.csv')
ITOsn = ITOs['n'].values 
ITOsk = ITOs['k'].values
ITOsingle = ml.Material(w, ITOsn, ITOsk, w)

ToS_hypo = tm.Design(
    [void, ITOsingle, SiO2, NB1, SiO2, NB1, sub],
    [None, 126, 23.8, 13.8, 55.6, 5.6, None]
).reflectance(wl, 5)

ToS2 = pd.read_csv('./spectrum/ToS2/ToS2.csv')['SCAN'].values
plt.plot(wl, ToS2, 'ks', markersize = 1)
plt.plot(wl, 100*ToS_hypo)
plt.show()
```
