```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import pi, sin, cos, dot, tensordot
from numpy.linalg import multi_dot
```

```python
class TF:
    def __init__(self, n, d):
        self.index = n
        self.thickness = d
        
    def matrix(self, w):
        delta = 2*pi*(self.index)*(self.thickness)/w
        ita = self.index
        element = pd.DataFrame(
            {'e1':cos(delta), 'e2':1j/ita*sin(delta), 'e3':1j*ita*sin(delta), 'e4':cos(delta)}
        )
        matrix = np.reshape(element.values.reshape(1,-1), (np.size(w), 2, 2))
        return matrix
```

```python
def FoM(R1, R0):
    M = sum(abs(R1-R0))/np.size(target_w)
    return M*100 #(%)
```

```python
target_w = np.linspace(430,680, 1000)
d_ITO  = 23 # (nm)
d_IM1 = 58 # (nm)
d_IM2 = 58 # (nm)
n_ITO = 1.9
n_IM1 = 1.89
n_IM2 = 1.91
TP_glass = 1.5 
cover_glass = 1.5 
```

```python
# spectrum
layer1 = TF(n_IM1, d_IM1).matrix(target_w) # IM1 
layer2 = TF(n_ITO, d_ITO).matrix(target_w) # ITO 
layer2_IM = TF(n_IM2, d_ITO).matrix(target_w) # IM2_match
layer3 = TF(n_IM2, d_IM2).matrix(target_w) # IM2
# R w/ ITO
Y0 = Eq_(layer3, Eq_(layer2, layer1))
R0 = R_(Y0, TP_glass, cover_glass)
# R w/o ITO
Y1 = Eq_(layer3, Eq_(layer2_IM, layer1))
R1 = R_(Y1, TP_glass, cover_glass)
plt.plot(target_w, R1, target_w, R0)
plt.show()
plt.plot(target_w, abs(R1-R0))
plt.show()
```

```python
# increase ITO
merit = []
for d_ITO in range(20, 1200):
    layer1 = TF(n_IM1, d_IM1).matrix(target_w) # IM1 
    layer2 = TF(n_ITO, d_ITO).matrix(target_w) # ITO 
    layer2_IM = TF(n_IM2, d_ITO).matrix(target_w) # IM2_match
    layer3 = TF(n_IM2, d_IM2).matrix(target_w) # IM2
    # R w/ ITO
    Y0 = Eq_(layer3, Eq_(layer2, layer1))
    R0 = R_(Y0, TP_glass, cover_glass)
    # R w/o ITO
    Y1 = Eq_(layer3, Eq_(layer2_IM, layer1))
    R1 = R_(Y1, TP_glass, cover_glass)
    merit.append(FoM(R1, R0))
    
plt.plot(np.linspace(20, 1200, np.size(merit)), merit)
plt.show()
```
