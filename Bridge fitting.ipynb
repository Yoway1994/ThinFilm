```python
import numpy as np
import pandas as pd
import thinfilm as tm
import material as ml
import matplotlib.pyplot as plt
```

```python
SiO2 = ml.open_material('SiO2', 'Sellmeier')
SiN = ml.open_material('SiN', 'Sellmeier')
ITO = ml.open_material('ITO', 'Sellmeier')
Acrylic = ml.open_material('Acrylic', 'Sellmeier')
OC = ml.open_material('OC', 'Sellmeier')
sodalime = ml.open_material('sodalime', 'Material')
air = ml.Non_Dispersion(1)
SiO2.k = 1e-3
ITO.k = 1e-1
SiN.k = 1e-1
```

```python
bridge = tm.Design([air, Acrylic, SiN, SiO2, ITO, OC, ITO, SiO2, SiN, sodalime],
                   [None, 1760, 10, 50, 23, 1500, 23, 50, 10, None])
```

```python
wl = np.linspace(400, 700, 301)
R = bridge.reflectance(wl)
plt.plot(wl, R)
plt.show()
```

```python
SiO2_t = ml.Non_Dispersion(1.46, 1e-5)
SiN_t = ml.Non_Dispersion(2.5, 6e-1)
ITO_t = ml.Non_Dispersion(1.9, 6e-1)
glass = ml.Non_Dispersion(1.52)
bridge2 = tm.Design([air, Acrylic, SiN_t, SiO2_t, ITO_t, OC, ITO_t, SiO2_t, SiN_t, sodalime],
                    [None, 1733, 8, 40, 28, 1200, 28, 60, 8, None])
R2 = bridge2.reflectance(wl)
plt.plot(wl, R2)
plt.savefig('bn.png')

plt.show()
print(np.mean(R2))
```
