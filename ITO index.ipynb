```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ITO = pd.read_csv('ITO_index.csv')
index = ITO[:381]
ext = ITO[382:]
wl_n = [float(i) for i in index['wl']]
n = [float(i) for i in index['n']]
wl_k = [float(i) for i in ext['wl']]
k = [float(i) for i in ext['n']]
```
