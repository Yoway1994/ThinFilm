```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DX1 = pd.read_csv('./DX/DX-MP-1_0T.csv')
GG3 = pd.read_csv('./DX/Glass_0T.csv')
wl = DX1['nm'][20:-100].values
DXT = DX1['SCAN'][20:-100].values/100
GGT = GG3['SCAN'][20:-100].values/100
DXR = 1-DXT
GGR = 1-GGT
SL_bri = pd.read_csv('./SE_spectrum/SL_bri.csv', index_col = 0)['R'].values
SL_dia = pd.read_csv('./SE_spectrum/SL_dia.csv', index_col = 0)['R'].values
SL_tra = pd.read_csv('./SE_spectrum/SL_tra.csv', index_col = 0)['R'].values
sig_bri_DX = abs(SL_bri-SL_dia)*DXT**2/DXR
sig_tra_DX = abs(SL_tra-SL_dia)*DXT**2/DXR
sig_bri_GG = abs(SL_bri-SL_dia)*GGT**2/GGR
sig_tra_GG = abs(SL_tra-SL_dia)*GGT**2/GGR
```

```python
plt.plot(wl, abs(SL_bri-SL_dia))
plt.plot(wl, abs(SL_tra-SL_dia))
plt.show()
```

```python
plt.plot(wl, sig_bri_DX)
plt.plot(wl, sig_tra_DX)
print(np.mean(sig_bri_DX))
print(np.mean(sig_tra_DX))
plt.show()
plt.plot(wl, sig_bri_GG)
plt.plot(wl, sig_tra_GG)
print(np.mean(sig_bri_GG))
print(np.mean(sig_tra_GG))
plt.show()
```
