```python
import pandas as pd
import matplotlib.pyplot as plt
CIE = pd.read_csv("CIE1931.csv")
CIE.columns = ["wl","R","G","B"]
plt.plot(CIE['wl'], CIE["R"], 'r')
plt.plot(CIE['wl'], CIE["G"], 'g')
plt.plot(CIE['wl'], CIE["B"], 'b')
plt.ylabel('a.u.')
plt.xlabel('wavelength')
plt.show()
```
