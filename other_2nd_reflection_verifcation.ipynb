```python
import numpy as np
import pandas as pd
import material as ml
import thinfilm as tm
import matplotlib.pyplot as plt

A = ml.Non_Dispersion(1)
G = ml.Non_Dispersion(1.5)
H = ml.Non_Dispersion(2.1)
L = ml.Non_Dispersion(1.4)
HH = ml.Non_Dispersion(5)

model = tm.Design(
    [A, L, H, G],
    [None, 100, 100, None]
)
wl = np.linspace(400, 700, 301)
angle = np.linspace(0,90,91)

R = model.reflectance(wl, 70, pl = 'avg')
T = model.transmittance(wl, 70, pl = 'avg')

plt.plot(wl, R)
plt.plot(wl, T)
plt.plot(wl, R+T)
plt.show()
```
