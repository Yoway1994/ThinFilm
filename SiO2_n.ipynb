```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
wavelength = [492.6, 535.88, 585.9, 643.34, 732.46, 838.07, 962.84, 1161.82, 1425.45]
wl = np.array(wavelength)/1000 # micron
SiO2_n = np.array([1.44538, 1.43662, 1.42846, 1.4226, 1.41846, 1.41575, 1.41422, 1.41375, 1.41369])
```

```python
def Cauchy(x):
    A = x[0]
    B = x[1]
    C = x[2]
    n = A + B/wl**2 + C/wl**4
    return sum(abs(n-SiO2_n))
kk = np.linspace(0.4, 2)
f = minimize(Cauchy, [0.1, 0.1, 0.1])
plt.plot(wl, SiO2_n, 'r-')
plt.plot(kk, [f.x[0] + f.x[1]/w**2 + f.x[2]/w**4 for w in kk])
plt.show()
```

```python
def Sellmeier(x):
    A = x[0]
    B = x[1]
    C = x[2]
    n = A + B/(1-C/wl**2)
    return sum(abs(n-SiO2_n**2))

ff = minimize(Sellmeier, [0.1, 0.1, 0.1])
plt.plot(wl, SiO2_n, 'r-')
plt.plot(kk, [(ff.x[0] + ff.x[1]/(1-ff.x[2]/w**2))**.5 for w in kk])
plt.show()
```

```python
# preprocessing
w_micron = np.array(wavelength)/1000
# training data
target_x = w_micron**2
target_y = np.array(SiO2_n)**2 - 1 
# sample number
m = np.size(target_x)
# initial value
A = 0.6961663
B = 0.1684043**2
C = 0.3079426
D =  0.1162414**2
E =  0.6974794
F =  11.896161**2
hofx = (A*target_x)/(target_x-B) + (C*target_x)/(target_x-D) + (E*target_x)/(target_x-F)
plt.plot(target_x, hofx)
plt.plot(target_x, target_y, '-s')
plt.show()
# cost function
testi = []
testJ = []
# learning rate
alpha = 0.001
# iteration
for i in range(5000):
    # hypothesis
    hofx = (A*target_x)/(target_x-B) + (C*target_x)/(target_x-D) + (E*target_x)/(target_x-F)
    # gradient descent
    Grad_J = 1/m*sum((hofx-target_y)*target_x.T)
    A = A - alpha* Grad_J
    B = B - alpha* Grad_J
    C = C - alpha* Grad_J
    D = D - alpha* Grad_J
    E = E - alpha* Grad_J
    F = F - alpha* Grad_J
    # cost function
    J = 1/(2*m)*sum((hofx-target_y)**2)
    testi.append(i)
    testJ.append(J)
# plot figure
plt.plot(testi, testJ)
plt.show()
print('cost =', testJ[-1])
plt.plot(target_x, hofx)
plt.plot(target_x, target_y, '-s')
plt.show()
print(A, B, C, D, E, F)
```
