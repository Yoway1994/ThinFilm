```python
import numpy as np
import matplotlib.pyplot as plt
wavelength = [492.6, 535.88, 585.9, 643.34, 732.46, 838.07, 962.84, 1161.82, 1425.45]
SiO2_n = [1.44538, 1.43662, 1.42846, 1.4226, 1.41846, 1.41575, 1.41422, 1.41375, 1.41369]
```

```python
# preprocessing
w_micron = np.array(wavelength)/1000
# training data
target_x = w_micron**2
target_y = np.array(SiO2_n)**2 - 1 
# sample number
m = np.size(target_x)
# initial value
A = 0.6961663
B = 0.1684043**2
C = 0.3079426
D =  0.1162414**2
E =  0.6974794
F =  11.896161**2
hofx = (A*target_x)/(target_x-B) + (C*target_x)/(target_x-D) + (E*target_x)/(target_x-F)
plt.plot(target_x, hofx)
plt.plot(target_x, target_y, '-s')
plt.show()
# cost function
testi = []
testJ = []
# learning rate
alpha = 0.001
# iteration
for i in range(5000):
    # hypothesis
    hofx = (A*target_x)/(target_x-B) + (C*target_x)/(target_x-D) + (E*target_x)/(target_x-F)
    # gradient descent
    Grad_J = 1/m*sum((hofx-target_y)*target_x.T)
    A = A - alpha* Grad_J
    B = B - alpha* Grad_J
    C = C - alpha* Grad_J
    D = D - alpha* Grad_J
    E = E - alpha* Grad_J
    F = F - alpha* Grad_J
    # cost function
    J = 1/(2*m)*sum((hofx-target_y)**2)
    testi.append(i)
    testJ.append(J)
# plot figure
plt.plot(testi, testJ)
plt.show()
print('cost =', testJ[-1])
plt.plot(target_x, hofx)
plt.plot(target_x, target_y, '-s')
plt.show()
print(A, B, C, D, E, F)
```
